```python
'''
4. ¿Cuántos pokemon tiene una defensa superior a 100? --> Hay 123 pokémon con una defensa superior a 100
'''
'''
4.1 ¿Cuántos pokemon tiene una defensa superior a 100?
'''
# Creo un nuevo dataframe con los pokémon que tenga una defena superior a 100
defense = df[df['Defense'] > 100]
# Reseteo los indices y elimino los anteriores con drop:
defense.reset_index(drop=True, inplace=True)
defense.head()
defense.tail()

# Es más rápido si directamente hago describe a defense y miro count. 
# Además, mirando tail tengo que sumarle 1 porque el índice empieza en 0.
defense.describe()
```

```python
'''
5. ¿Cuántos pokemon tiene una defensa superior a 100 y una velocidad superior a 100? --> Hay 9 pokémon
'''
# Creo un filtro con los pokemon con defensa > 100
defense100 = df['Defense'] > 100
# Creo un filtro con los pokemon con velocidad > 100
speed100 = df['Speed'] > 100
# Creo un dataframe con los pokémon que cumplen ambos filtros: 
defensespeed100 = df[defense100 & speed100]
# Reseteo los indices y elimino los anteriores con drop:
defensespeed100.reset_index(drop=True, inplace=True)
defensespeed100.tail()

# Es mucho más rápido aplicar directamente un count:
defensespeed100.count()
```
